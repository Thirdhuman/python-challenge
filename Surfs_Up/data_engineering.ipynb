```python
# Dependencies
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
csv_path1='Resources/hawaii_measurements.csv'
meas_df = pd.read_csv(csv_path1)
csv_path2='Resources/hawaii_stations.csv'
station_df = pd.read_csv(csv_path2)

print(meas_df.dtypes)
print(meas_df.dtypes)

print(meas_df.columns.values)
print(station_df.columns.values)

print(station_df['name'].unique)

print(meas_df.head())
print(station_df.head())

print(pd.isnull(station_df))
print(pd.isnull(meas_df))
```

```python
print(meas_df.count())
#meas_df= meas_df[pd.notnull(meas_df)]
meas_df=meas_df.dropna()
print(meas_df.count())

meas_df['date']=  pd.to_datetime(meas_df['date'], format='%Y-%m-%d')

#Between 2016/08/23 to 2017/08/23

meas_df=meas_df[(meas_df['date'] > '2016-08-23') & (meas_df['date'] < '2017-08-23')]
print(meas_df.count())

meas_df.head(5)
```

```python
print(station_df.count())
station_df['name'] = station_df['name'].str.replace(', HI US', '')
station_df
```

```python
station_df.to_csv('resources/clean_station_df.csv', encoding='utf-8', index=False)
meas_df.to_csv('resources/clean_meas_df.csv', encoding='utf-8', index=False)
```
